```python
RandomForestClassifier?
```

https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

**n_estimators (int, default=100) =** número de árvores na floresta

**bootstrap (bool, default=True) =** True significa que, em vez de treinar em todas as observações, cada árvore é treinada em um subconjunto das observações. Várias árvores são treinadas em diferentes grupos e, posteriormente, os resultados de todas as árvores são agregados. Se for False todo o conjunto de dados é usado para construir cada árvore.

**max_samples (int ou float, default=Nenhum)** : Define o número de amostras para treinamento de cada grupo quando o bootstrap for True.
